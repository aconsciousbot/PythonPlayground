## Breadth First Search

```python
def breadth_first_traversal(root, graph):
  """
  Given a node (amenable to comparison and sorting) and a graph (dict-like), 
  returns a list of nodes that are reached from root
  BFS uses a FIFO approach, using list-based Queue here
  """
  visited = list()
  Q = list()
  Q.append(root)

  while len(Q)>0:
    #print(Q, visited)
    node_current = Q.pop(0)

    if node_current not in visited:
      visited.append(node_current)
      nodes_adjacent = graph[node_current]
      for node in sorted(nodes_adjacent):
        if node not in visited and node not in Q:
          Q.append(node)
  return visited
```

```python
breadth_first_traversal("A", graph)
```

## Depth First Search

```python
def depth_first_traversal(root, graph):
  """
  Given a node (amenable to comparison and sorting) and a graph (dict-like), 
  returns a list of nodes that are reached from root
  DFS uses a LIFO approach, using a list-based Stack here
  """
  visited = list()
  stack = list()
  stack.append(root)

  while len(stack)>0:
    #print(stack, visited)
    node_current = stack.pop()

    if node_current not in visited:
      visited.append(node_current)
      nodes_adjacent = graph[node_current]
      for node in sorted(nodes_adjacent)[::-1]:
        if node not in visited:
          stack.append(node)
                  
          
  return visited
```

```python
depth_first_traversal("A", graph)
```
